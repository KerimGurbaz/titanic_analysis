```python
Q1= df["Fare"].quantile(0.25)
```

```python
Q3 = df["Fare"].quantile(0.75)
```

```python
IQR = Q3-Q1
```

```python
IQR
```

```python
lower_bound= Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
```

```python
outliers = df[(df["Fare"] < lower_bound) | (df["Fare"] > upper_bound)]
```

```python
df["Fare"]= df["Fare"].apply(lambda x : x if lower_bound<= x <=upper_bound else np.nan)
```

```python
df[df["Fare"].isnull()]
```

```python
nan_count = df['Fare'].isnull().sum()
print(f"Fare stununda {nan_count} tane NaN deeri var.")
```

```python
df['Fare'] = df['Fare'].fillna(df['Fare'].mean())
```

```python
nan_count_after = df['Fare'].isnull().sum()
print(f"Doldurma ileminden sonra Fare stununda {nan_count_after} tane NaN deeri var.")
```
